```python
#Read the three files into python as dataframes
import pandas as pd
import numpy as np

file1 = pd.read_csv('/Users/ayeesha/IH-Labs/lab-customer-analysis-round-1/files_for_lab/csv_files/file1.csv')
file2 = pd.read_csv('/Users/ayeesha/IH-Labs/lab-customer-analysis-round-1/files_for_lab/csv_files/file2.csv')
file3 = pd.read_csv('/Users/ayeesha/IH-Labs/lab-customer-analysis-round-1/files_for_lab/csv_files/file3.csv')
display(file1.shape)
display(file2.shape)
display(file3.shape)




cols = []
for column in file1.columns:
    cols.append(column.lower())
file1.columns = cols



file1.rename(columns = {'st':'state'}, inplace = True)

display(file1.head())
```

```python

cols = []
for column in file2.columns:
    cols.append(column.lower())
file2.columns = cols



file2.rename(columns = {'st':'state'}, inplace = True)

display(file2.head())
```
